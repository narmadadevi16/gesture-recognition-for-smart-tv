```python
data=[]
labels=[]

for images in pathToImages:
    label=images.split(os.path.sep)[-2] #contains all the label names
    
    label=label[26:]
    label=label[:-4]
    
    if label[:4]=="Left":
        label="Left Swipe"
    if label[:5]=="Right":
        label="Right Swipe"
    if label[:4]=="Stop":
        label="Stop Gesture"
    if label[-4:]=="Down":
        label="Thumbs Down"
    if label[-2:]=="Up":
        label="Thumbs Up"
    

    image=cv2.imread(images)  #returns a NumPy array representing the image. The array contains the pixel values of the image.
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    image=cv2.resize(image,(360,360))
    data.append(image)
    labels.append(label)
```

```python
set(labels)
```

```python
data=np.array(data)
labels=np.array(labels)


#one hot encoding

lb=LabelBinarizer()
labels=lb.fit_transform(labels)
```

```python

# Reverse the one-hot encoding to obtain the original labels
original_labels = lb.inverse_transform(labels)
```

```python
(X_train,X_test,Y_train,Y_test)=train_test_split(data,labels,test_size=0.25,stratify=labels,random_state=0)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Create a sequential model
model = Sequential()

# Add the first convolutional layer
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(360, 360, 3)))

# Add a max pooling layer
model.add(MaxPooling2D(pool_size=(2, 2)))

# Add a second convolutional layer
model.add(Conv2D(64, (3, 3), activation='relu'))

# Add another max pooling layer
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten the 3D feature maps to 1D
model.add(Flatten())

# Add a fully connected layer
model.add(Dense(64, activation='relu'))

# Add the output layer with the appropriate number of units/classes
model.add(Dense(5, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()
```

```python
# Assuming you have your training data and labels stored in variables: train_data and train_labels

# Fit the model to the training data
model.fit(X_train, Y_train, epochs=2, batch_size=32)
```

```python
# Assuming you have the test data and labels stored in variables: test_data and test_labels
test_loss, test_accuracy = model.evaluate(X_test, Y_test)
print("Test Accuracy:", test_accuracy)
```
