```python
import cv2
import numpy as np

label_mapping={0:'Left Swipe',1:'Right Swipe',2:'Stop',3:'Thumbs Up',4:'Thumbs Down'}
cap = cv2.VideoCapture(0)
window_size = 5  # Adjust the window size as needed

# List to store predicted labels from each frame
predictions = []

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break
    out=frame.copy()
        
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    frame = cv2.resize(frame, (360, 360))
    
    eframe = np.expand_dims(frame, axis=0)

    # Preprocess the frame if required
    # ...

    # Make predictions for the current frame
    current_prediction = model.predict(eframe)  # Replace with your prediction code

    # Store the current prediction in the list
    predictions.append(current_prediction)

    # Perform temporal smoothing using a sliding window
    if len(predictions) > window_size:
        # Compute the average prediction over the sliding window
        smoothed_prediction = np.mean(predictions[-window_size:], axis=0)

        # Get the predicted class index with the highest probability
        predicted_class_index = np.argmax(smoothed_prediction)

        # Map the predicted class index to the corresponding label
        predicted_label = label_mapping[predicted_class_index]  # Replace with your label mapping

        # Overlay the label text on the frame
        cv2.putText(out, predicted_label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Display the frame
    cv2.imshow('Video', out)

    # Check for key press events
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
